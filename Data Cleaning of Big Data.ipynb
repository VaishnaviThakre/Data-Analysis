```python
import pandas as pd

summer = pd.read_csv('summer.csv')

# intial dataset
print("Initial Dataset Information:")
print(summer.info())

# Handling missing values
print("\nHandling Missing Values:")
print("Number of missing values per column:")
print(summer.isnull().sum())  

summer['Country'].fillna('Unknown', inplace=True)

# Removing duplicates
print("\nRemoving Duplicates:")
initial_rows = summer.shape[0]  
summer.drop_duplicates(inplace=True)
final_rows = summer.shape[0]
print(f"Initial number of rows: {initial_rows}")
print(f"Final number of rows after removing duplicates: {final_rows}")

# Handling inconsistent data (converting strings to proper case)
cols_to_clean = ['City', 'Sport', 'Discipline', 'Athlete', 'Country', 'Gender', 'Event', 'Medal']
for col in cols_to_clean:
    summer[col] = summer[col].str.title()  # Convert strings to proper case (title case)

print("\nCleaned Dataset Information:")
print(summer.info())

#cleaned_summer.csv'
summer.to_csv('cleaned_summer.csv', index=False)
```
