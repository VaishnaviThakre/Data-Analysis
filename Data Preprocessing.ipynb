# iloc
iloc is primarily used for selecting rows and columns in a Pandas DataFrame by their integer position or index.

It uses integer-based indexing, meaning you specify the row and column positions by their numerical index rather than by labels.

use iloc to slice, select specific rows/columns, or access specific elements in a DataFrame.

It takes two integers separated by a comma, the first integer represents the row positions, and the second integer represents the column positions.

```python
sum1 = summer.iloc[1:30]
sum1.head(10)
```

```python
sum2= summer.iloc[40:100]
sum2.head(10)
```

# isNull 
This method is used to detect missing or NaN (Not a Number) values in a DataFrame or Series.

When applied to a DataFrame or Series, isnull() returns a Boolean mask (True/False) indicating whether each element is missing (NaN) or not.

It returns True for positions where the value is NaN and False otherwise.

```python
sum1.isnull()
```

# notnull
This method is the opposite of isnull() and is used to detect non-missing values in a DataFrame or Series.

It returns False for positions where the value is NaN and True otherwise.

```python
sum1.notnull()
```

# dropna
This Pandas function is applied to a DataFrame or Series to remove rows or columns that contain any missing values.
By default, it removes rows (axis=0) with any NaN values.
It returns a new DataFrame or Series without the rows or columns containing missing values.

```python
sum2.dropna()
sum2.head(10)
```

# fillna
fillna() is a method in Pandas used to replace NaN values with a specified value or a predefined strategy.

It can be applied to a DataFrame or Series.

```python
sum1.fillna("--")
sum1.head(10)
```

# replace

replace() is a method in Pandas used to replace values in a DataFrame or Series with other values.

It can handle various replacements, such as replacing a single value, multiple values, or using a dictionary to specify replacements for different values.

```python
summer.replace("Men","Male")
```

# interpolate
interpolate() is a method in Pandas used to fill NaN values in a DataFrame or Series by performing interpolation between existing values.

It can handle different types of interpolation methods such as linear, quadratic, cubic, etc.

```python
sum2 = summer.interpolate(method="linear")
sum2
```

# filter data
Filtering data in Pandas involves selecting rows or columns based on specific conditions or criteria.

```python
filtered_data = summer[['Year', 'City', 'Sport', 'Discipline', 'Athlete', 'Country', 'Gender', 'Event', 'Medal']]
print(filtered_data.head())
```

# creating dataframe using dict
To create DataFrame from dict of narray/list, all the narray must be of same length. If index is passed then the length index should be equal to the length of arrays. If no index is passed, then by default, index will be range(n) where n is the array length.

```python
data = {'Name': ['Tom', 'nick', 'krish', 'jack'],
'Age': [20, 21, 19, 18]}

df = pd.DataFrame(data)

df
```
